## Training for color

```python
X = df.drop(['BEST COLOR','BEST POSITION','DEVICE'],axis=1).copy()
y = df['BEST COLOR'].copy()
```

```python
# X.drop(['GENDER','GLASSES','POSITION'],axis=1,inplace=True)
x = pd.get_dummies(df[['GENDER','GLASSES','POSITION']],drop_first=True)
X = pd.concat([X,x],axis=1)
X.drop(["GLASSES","POSITION","GENDER"],axis=1,inplace=True)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state=101)
```

```python
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
from sklearn.metrics import accuracy_score
models = [LogisticRegression(),SVC(),KNeighborsClassifier(),DecisionTreeClassifier(),RandomForestClassifier()]
def validate_models(models,X_train,y_train,X_test,y_test):
    ind = []
    scores = []
    for model in models:
        model = model.fit(X_train,y_train)
        preds = model.predict(X_test)
        acc_score = accuracy_score(y_test,preds)
        scores.append([acc_score])
        ind.append(str(model))
    vdf = pd.DataFrame(index=ind,columns=['SCORE1'],data=scores)
    return vdf
print(validate_models(models=models,X_train=X_train_sc,y_train=y_train,X_test=X_test_sc,y_test=y_test))
```

```python
ddf = pd.get_dummies(df,drop_first=True)
```

```python
sc = StandardScaler()
ddf_sc = sc.fit_transform(ddf)
```

```python
from sklearn.cluster import KMeans
sse = []
for i in range(2,6):
    model = KMeans(n_clusters=i)
    model.fit(ddf_sc)
    sse.append(model.inertia_)
```

```python
plt.plot([2,3,4,5],sse,marker='o',linestyle='--')
plt.xlabel('Clusters')
plt.ylabel('Sum of Squared error')
plt.savefig('cluster.png')
```
